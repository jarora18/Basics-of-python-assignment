```python
#Q6
import random as r
L = [r.randint(100, 900) for _ in range(100)]
print(L)
#(i)
for i in L:
  if i%2!=0:
    print(i)
#(ii)
for i in L:
  if i%2==0:
    print(i)
#(iii)
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
prime_numbers = [num for num in L if is_prime(num)]
print(prime_numbers)
```

```python
#Q10
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

fp= open("prime_600_800.txt", "w")
for num in range(600, 801):
        if is_prime(num):
            fp.write(str(num) + "\n")
fp.close()
```

```python
#Q12
import time
import random
import matplotlib.pyplot as plt
sizes = [5000, 10000, 15000, 20000, 25000]
times = []
for size in sizes:
    data = [random.randint(1, 100000) for _ in range(size)]
    start_time = time.time()
    sorted_data = sorted(data)
    end_time = time.time()
    time_taken = end_time - start_time
    times.append(time_taken)
    print(f"Sorted list of {size} elements in {time_taken:.6f} seconds.")


plt.plot(sizes, times, marker='o', color='b')
plt.title("Time Taken to Sort Lists of Different Sizes")
plt.xlabel("Number of Elements in List")
plt.ylabel("Time Taken (seconds)")
plt.grid(True)
plt.show()
```

```python
#Q13
student_marks = {
    "Alice": [78, 85, 92, 88, 76],
    "Bob": [67, 74, 81, 69, 73],
    "Charlie": [90, 91, 89, 94, 88],
    "David": [56, 60, 52, 58, 54],
    "Eva": [80, 85, 83, 87, 82]
}
total_marks = {}
for student, marks in student_marks.items():
    total = sum(marks)
    total_marks[student] = total

max_student = max(total_marks, key=total_marks.get)
min_student = min(total_marks, key=total_marks.get)

print("Total Marks of All Students:")
for student, total in total_marks.items():
    print(f"{student}: {total}")

print("\nStudent with Maximum Marks:")
print(f"{max_student} => {total_marks[max_student]}")

print("\nStudent with Minimum Marks:")
print(f"{min_student} => {total_marks[min_student]}")
```
